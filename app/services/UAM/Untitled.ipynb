```python
#to check the delete function
def check_deletion():
    from selenium.webdriver.common.action_chains import ActionChains
    rows_before = driver.find_elements_by_tag_name("tr")
    print("rows before deletion = "+str(len(rows_before)))
    element_to_hover_over = driver.find_element_by_xpath('//*[@id="root"]/div[3]/div/div/div[3]/div/div/div/table/tbody/tr[1]/td[2]/div')
    hover = ActionChains(driver).move_to_element(element_to_hover_over)
    hover.perform()
    driver.find_element_by_xpath('//*[@id="root"]/div[3]/div/div/div[3]/div/div/div/table/tbody/tr[1]/td[2]/div/button[1]').click()
    time.sleep(7)
    rows_after = driver.find_elements_by_tag_name("tr")
    print("rows after deletion = "+str(len(rows_before)))
    if len(rows_before)-len(rows_after) == 1:
        return ("deletion successful")
    else:
        return ("deletion failed")
```

```python
def get_filter_values_and_id(xpath):
    res = []
    driver.find_element_by_xpath(f'{xpath}/div/button[2]').click()
    element = driver.find_element_by_xpath(f'{xpath}/input')
    val = element.get_attribute('id')
    i = 0
    while True:
        result = {}
        try:
            element = driver.find_element_by_xpath(f'//*[@id="{val}-option-{i}"]/span/span/input')
            
            result['id'] = f'{val}-option-{i}'
            result['val'] = driver.find_element_by_xpath(f'//*[@id="{val}-option-{i}"]').text
            res.append(result)
            i += 1
        except Exception as e:
            print(e)
            break
    return res
```
